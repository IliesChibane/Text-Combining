```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Google': [0.885, 0.514, 0.825, 0.645, 0.861, 0.958, 0.952, 0.955, 0.755, 0.241],
    'Microsoft': [0.870, 0.602, 0.857, 0.697, 0.857, 0.968, 0.961, 0.964, 0.791, 0.168],
    'OpenAI': [0.903, 0.594, 0.892, 0.706, 0.892, 0.979, 0.977, 0.978, 0.813, 0.18],
    'AdvancedGenius': [0.905, 0.552, 0.885, 0.688, 0.869, 0.97, 0.968, 0.969, 0.811, 0.219]
}

labels = ['Exactitude', 'BLEU', 'ROUGE-1', 'ROUGE-2', 'ROUGE-L', 'BERTScore Precision', 'BERTScore Rappel', 'BERTScore F-Score', 'METEOR', 'WER']
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False)

angles = np.concatenate((angles,[angles[0]]))

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, polar=True)

for key in data:
    values = data[key]
    values = np.concatenate((values,[values[0]]))
    ax.plot(angles[:-1], values[:-1], linewidth=1, linestyle='solid', label=key)
    ax.fill(angles[:-1], values[:-1],alpha=0.1)

ax.set_thetagrids(angles[:-1] * (180/np.pi), labels)
plt.legend(loc='upper right', bbox_to_anchor=(1.3 ,1))

plt.savefig('radar.png')
plt.show()
```

```python
import plotly.graph_objects as go

data = {
    'Exactitude': 0.916,
    'BLEU': 0.661,
    'ROUGE-1': 0.904,
    'ROUGE-2': 0.767,
    'ROUGE-L': 0.890,
    'BERTScore Precision': 0.949,
    'BERTScore Rappel': 0.958,
    'BERTScore F-Score': 0.953,
    'METEOR': 0.878,
    'WER': 0.129
}

fig = go.Figure()

fig.add_trace(go.Waterfall(
    name = "Premire approche",
    orientation = "v",
    measure = ["relative"]*len(data),
    x = list(data.keys()),
    textposition = "outside",
    text = [f"{value:.3f}" for value in data.values()],
    y = list(data.values()),
    connector = {"line":{"color":"rgb(63, 63, 63)"}},
))

fig.update_layout(
        title = "Contribution de chaque mtrique  la performance globale de notre premire approche",
        showlegend = True
)

fig.show()
```
